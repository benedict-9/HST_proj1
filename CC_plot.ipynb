### Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Read CSV File

```python
df = pd.read_csv("CommunityClubsTest-clean.csv")
```

### Plot

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['X'], df['Y'], color='blue', marker='o')  # You can customize the marker and color
plt.xlabel("Longitude (x)")
plt.ylabel("Latitude (y)")
plt.title("XY Coordinate Plot")
plt.grid(True)
plt.show()
```
